```python
from rapidfuzz import process, fuzz

def find_similar_names(dataframe, column_name, threshold=80):

    unique_names = dataframe[column_name].unique()
    similar_names = []

    for name in unique_names:
        matches = process.extract(name, unique_names, scorer=fuzz.ratio)
        for match in matches:
            if match[1] >= threshold and match[0] != name:
                similar_names.append((name, match[0], match[1]))
    return similar_names

similar_names = find_similar_names(d1, 'pc_name', 80)

for name_pair in similar_names:
    print(f"{name_pair[0]} <-> {name_pair[1]} (Similarity: {name_pair[2]})")
```

```python
def find_similar_names(dataframe, column_name, threshold=80):
    unique_names = dataframe[column_name].unique()
    similar_names = []

    for name in unique_names:
        matches = process.extract(name, unique_names, scorer=fuzz.ratio)
        for match in matches:
            if match[1] >= threshold and match[0] != name:
                similar_names.append((name, match[0], match[1]))

    return similar_names

# Function to get constituencies with spelling mismatches
def get_mismatched_constituencies(dataframe):
    mismatched_constituencies = []

    grouped = dataframe.groupby(['state', 'total_electors'])
    for _, group in grouped:
        similar_names = find_similar_names(group, 'pc_name', 80)
        if similar_names:
            mismatched_constituencies.extend(group['pc_name'].unique())
    
    return mismatched_constituencies

# Get constituencies with potential spelling mismatches
mismatched_constituencies = get_mismatched_constituencies(d1)

print("Constituencies with potential spelling mismatches:")
print(mismatched_constituencies)
```

```python
def analyse_voter_turnout(d1):
    total_votes_per_constituencies = d1.groupby('pc_name')['total_votes'].sum().reset_index()
    total_electors_per_constituencies = d1[['pc_name', 'total_electors']].drop_duplicates()
    
    
    merged_df = pd.merge(total_votes_per_constituencies, total_electors_per_constituencies, on='pc_name')
    
    merged_df['voter_turnout_ratio'] = (merged_df['total_votes']/merged_df['total_electors'])*100
    
    sorted_voter_turnout = merged_df.sort_values(by='voter_turnout_ratio', ascending=False)
    
    top5_constituencies = sorted_voter_turnout.head(5)
    bottom5_constituencies = sorted_voter_turnout.tail(5)

    return top5_constituencies,bottom5_constituencies

top_2014, bottom_2014 = analyse_voter_turnout(d1)
top_2019, bottom_2019 = analyse_voter_turnout(d2)

print(f"top 5 constituencies in terms of their voter turnout ratio in the year 2014 \n\n {top_2014} \n\n and bottom 5 are \n\n {bottom_2014}\n\n")
print(f"top 5 constituencies in terms of their voter turnout ratio in the year 2019 \n\n {top_2019} \n\n and bottom 5 are \n\n {bottom_2019}")
```

```python
def states_voter_turnout_ratio(d1):

    total_electors_per_state = d1.groupby('state')['total_electors'].sum().reset_index()
    total_votes_per_state = d1.groupby('state')['total_votes'].sum().reset_index()

    merge_df = pd.merge(total_electors_per_state, total_votes_per_state, on='state')

    merge_df['voter_turnout_state'] = (merge_df['total_votes']/merge_df['total_electors']) * 100

    sorted = merge_df.sort_values(by='voter_turnout_state', ascending=False)

    top5_states = sorted.head(5)
    bottom5_states = sorted.tail(5)

    return top5_states, bottom5_states
    
```

```python
top5_states2014, bottom5_states2014 = states_voter_turnout_ratio(d1)

top5_states2019, bottom5_states2019 = states_voter_turnout_ratio(d2)

print(f"top 5 states in terms of their voter turnout ratio in the year 2014 \n\n {top5_states2014} \n\n and bottom 5 are \n\n {bottom5_states2014}\n\n")
print(f"top 5 constituencies in terms of their voter turnout ratio in the year 2019 \n\n {top5_states2019} \n\n and bottom 5 are \n\n {bottom5_states2019}")
```

```python
def find_winning_party(d1):

    winning_candidates = d1.loc[d1.groupby('pc_name')['total_votes'].idxmax()].reset_index(drop=True)
    
    winning_parties = winning_candidates[['state', 'pc_name', 'candiwdate', 'party', 'total_votes']]

    return winning_parties

def analyze_consecutive_winners(d1, d2):
    winning_2014 = find_winning_party(d1)
    winning_2019 = find_winning_party(d2)
    
    merged = pd.merge(winning_2014, winning_2019, on='pc_name', suffixes=('_2014', '_2019'))
    same_party = merged[merged['party_2014'] == merged['party_2019']]

    same_party['voter_percentage_2019'] = (same_party['total_votes_2019']/ d2.groupby('pc_name')['total_votes'].sum()[same_party['pc_name']].values)* 100

    same_party_sorted = same_party.sort_values(by='voter_percentage_2019', ascending=False)

    return same_party_sorted[['state_2019', 'pc_name', 'party_2019', 'voter_percentage_2019']]


same_party_sorted = analyze_consecutive_winners(d1, d2)

print("Constituencies that elected the same party for two consecutive elections, ranked by % votes to the winning party in 2019:\n")
print(same_party_sorted)
```

```python
def analyse_votes_diff_of_parties(d1, d2):
    winning_2014 = find_winning_party(d1)
    winning_2019 = find_winning_party(d2)
    
    merged = pd.merge(winning_2014, winning_2019, on='pc_name', suffixes=('_2014', '_2019'))
    same_party = merged[merged['party_2014'] != merged['party_2019']]

    same_party['voter_percentage_2019'] = (same_party['total_votes_2019']/ d2.groupby('pc_name')['total_votes'].sum()[same_party['pc_name']].values)* 100
    same_party['voter_percentage_2014'] = (same_party['total_votes_2014']/ d1.groupby('pc_name')['total_votes'].sum()[same_party['pc_name']].values)* 100
    same_party['diff'] = same_party['voter_percentage_2019'] - same_party['voter_percentage_2014']
    
    same_party_sorted = same_party.sort_values(by='diff', ascending=False)

    return same_party_sorted[['state_2019', 'pc_name', 'party_2019', 'diff']]

df = analyse_votes_diff_of_parties(d1, d2)
df.head(10)
```

```python
def find_top_margin_candidates(dataframe, year):
    # Sort by constituency and total votes in descending order to get the winner and runner-up
    sorted_df = dataframe.sort_values(by=['pc_name', 'total_votes'], ascending=[True, False])
    
    # Group by constituency and take the top 2 candidates (winner and runner-up)
    top_two = sorted_df.groupby('pc_name').head(2)
    
    # Rank within each group to distinguish winner (rank 1) and runner-up (rank 2)
    top_two['rank'] = top_two.groupby('pc_name')['total_votes'].rank(method='first', ascending=False)
    
    # Pivot the dataframe to get winner and runner-up in separate columns
    pivot_df = top_two.pivot(index='pc_name', columns='rank', values=['candidate', 'party', 'total_votes'])
    pivot_df.columns = [f"{x}_{int(y)}" for x, y in pivot_df.columns]

    # Calculate the margin difference
    pivot_df['margin_difference'] = pivot_df['total_votes_1'] - pivot_df['total_votes_2']
    
    # Sort by margin difference and select top 5
    top_margin_candidates = pivot_df.sort_values(by='margin_difference', ascending=False).head(5)
    
    # Add the year for reference
    top_margin_candidates['year'] = year
    
    return top_margin_candidates.reset_index()
```

```python
top_margin_2014 = find_top_margin_candidates(d1, 2014)
top_margin_2019 = find_top_margin_candidates(d2, 2019)

top_margin_candidates = pd.concat([top_margin_2014, top_margin_2019])
print("Top 5 candidates based on margin difference with runners-up in 2014 and 2019:\n")
print(top_margin_candidates[['pc_name', 'candidate_1', 'party_1', 'total_votes_1', 'candidate_2', 'party_2', 'total_votes_2', 'margin_difference', 'year']])
```

```python
def calculate_party_vote_percentage(df, year):
    # Group by party and sum total votes
    party_votes = df.groupby('party')['total_votes'].sum().reset_index()
    
    # Calculate total votes cast in the year
    total_votes = party_votes['total_votes'].sum()
    
    # Calculate percentage of votes for each party
    party_votes['percent_votes'] = (party_votes['total_votes'] / total_votes) * 100
    
    # Add year for reference
    party_votes['year'] = year
    
    return party_votes

# Calculate percentage of votes for each party in 2014 and 2019
party_votes_2014 = calculate_party_vote_percentage(d1, 2014)
party_votes_2019 = calculate_party_vote_percentage(d2, 2019)

# Merge the results for comparison
merged_votes = pd.merge(party_votes_2014, party_votes_2019, on='party', suffixes=('_2014', '_2019'))

# Display the results
print("Percent split of votes between parties in 2014 and 2019 at the national level:\n")
print(merged_votes[['party', 'percent_votes_2014', 'percent_votes_2019']])
```
